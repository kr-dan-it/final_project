```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
sns.set_context("notebook")
sns.set_palette(sns.color_palette("magma_r"))
```

## 1. Аналіз та підготовка даних

Uploading the extended iris dataset from kaggle, which is like the Iris dataset but has more columns and more items (1200 irises in total): https://www.kaggle.com/datasets/samybaladram/iris-dataset-extended

```python
extended_iris = pd.read_csv('iris_extended.csv')
extended_iris.head()
```

Looking at the total number of columns and dropping some columns I won't be working with

```python
extended_iris.info()
```

```python
extended_iris_1 = extended_iris.drop(columns=['sepal_aspect_ratio', 'petal_aspect_ratio', 
                                              'sepal_to_petal_length_ratio', 'sepal_to_petal_width_ratio', 
                                             'sepal_petal_length_diff', 'sepal_petal_width_diff', 
                                             'petal_texture_trichomes_per_mm2', 'sepal_area_sqrt', 
                                              'petal_area_sqrt', 'area_ratios'])
extended_iris_1.head()
```

Describing the dataset, separately for numeric and categorical values

The only two categorical values are the iris species and the soil type

```python
extended_iris_1.describe()
```

```python
extended_iris_1.describe(include='object')
```

Looking at the amount of NaN values in the dataset. This one had 0

```python
extended_iris_1.isna().sum()
```

Doing label encoder on the two categorical values: first on the species, then on the soil type

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(extended_iris_1['species'])
species_code = le.transform(extended_iris_1['species'])
species_code
```

```python
extended_iris_1.insert(loc = 0,
          column = 'species_code',
          value = species_code)
```

```python
extended_iris_1.head()
```

```python
le.fit(extended_iris_1['soil_type'])
soil_code = le.transform(extended_iris_1['soil_type'])
soil_code
```

```python
extended_iris_1.insert(loc = 1,
          column = 'soil_code',
          value = soil_code)
```

```python
extended_iris_1.head()
```

Dropping the original species and soil type columns, since now we have the categorical columns added to the dataset as well

```python
extended_iris_2 = extended_iris_1.drop(columns=['species', 'soil_type'])
extended_iris_2.head()
```

Exporting extended_iris_2 into a csv file so I could use it later

```python
extended_iris_2.to_csv('edited_extended_iris_df.csv', index=False)
```

Building a correlation matrix

```python
correlation = extended_iris_2.iloc[:, :].corr()
correlation
```

```python
sns.heatmap(correlation, annot=True, fmt=".2f", vmin=-1, vmax=1)
```

Species correlate the most with petal_length and petal_area (0.94 each), petal_width (0.93) and petal_curvature (0.91)

Species do not correlate at all with the soil type (0.04) and the elevation (0.00)

## **2. Розробка моделі машинного навчання**

Getting dummy variables out of the soil codes, then renaming the dummy variable columns so it was easier to understand which number means what

```python
dummies = pd.get_dummies(extended_iris_2.soil_code)
dummies.head()
```

```python
extended_iris_1['soil_type'].unique()
```

```python
# sandy clay sandy clay loamy
```

```python
dummies.rename(columns={0: 'clay', 1: 'loamy', 2: 'sandy'}, inplace=True)
dummies.head()
```

```python
dummies['clay']
```

Adding the dummy columns to the dataset and dropping the initial soil_codes column. Also dropping the first dummy column, which is clay

```python
extended_iris_3 = pd.concat([extended_iris_2, dummies.drop('clay', axis='columns')], axis='columns')
extended_iris_3.head()
```

```python
extended_iris_3.drop(columns='soil_code', inplace=True)
extended_iris_3.head()
```

Creating the X and the y parts of the dataset

```python
X = extended_iris_3.drop('species_code', axis='columns')
X.head()
```

```python
y = extended_iris_3.species_code
y
```

Doing train test split, and then displaying the training and testing datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 19)
```

```python
X_train.head()
```

```python
X_test.head()
```

Running random forest classifier (I didn't run Pipeline because it's too taxing for my computer)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train.values, y_train.values)
```

```python
y_pred = rf.predict(X_test)
```

Looking at the score

```python
rf.score(X_test, y_test)
```

Getting cross validation scores

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
cvscore = cross_val_score(rf, X, y.values, cv=10)
cvscore
```

Doing a test prediction

```python
iris_type = rf.predict([[169, 5, 2, 3, 1, 10, 3, 8, 70, True, False]])
```

```python
iris_type
```

Doing hyper parameter tuning using Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = [{
    'n_estimators': [500, 1000, 1500], 
    'max_depth': [10, 20, 30], 
    'min_samples_split': [5, 10, 15], 
    'min_samples_leaf': [1, 2, 4]
}]
```

```python
random_grid_search = RandomizedSearchCV(rf, parameters, cv = 5, scoring='accuracy', n_jobs = -1, random_state = 19)
```

```python
random_grid_search.fit(X_train, y_train)
```

```python
random_grid_search.best_score_
```

```python
random_grid_search.best_params_
```

Creating a new Random Forest Classifier model, this time using the best parameters the random grid search gave us

```python
rf_tuned = RandomForestClassifier(n_estimators=1000, min_samples_split=5, min_samples_leaf=1, max_depth=30)
```

```python
rf_tuned.fit(X_train, y_train)
```

```python
y_pred_tuned = rf_tuned.predict(X_test)
```

Looking at the CV scores of the new RFC model, this time including average and std as well

```python
cvscore = cross_val_score(rf_tuned, X, y.values, cv=10)
cvscore
```

```python
average_cvscore = np.mean(cvscore)
average_cvscore
```

```python
std_cvscore = np.std(cvscore)
std_cvscore
```

Making a test prediction with the new RFC model

```python
iris_type_tuned = rf_tuned.predict([[169, 5, 2, 3, 1, 10, 3, 8, 70, True, False]])
iris_type_tuned
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy = accuracy_score(y_test, y_pred_tuned)
accuracy
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
c_matrix = confusion_matrix(y_test, y_pred_tuned)
c_matrix
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_tuned)
plt.grid(False)
plt.show()
```

The confusion matrix shows that the model made 57 predictions about Species 0 (Setosa), all of which were correct

The model made 84 correct predictions about Species 1 (Versicolor) but also 8 incorrect ones labeled as Virginica

The model also made 98 correct predictions about Species 2 (Virginica) but also 2 incorrect ones labeled as Versicolor

```python
print(classification_report(y_test, y_pred_tuned))
```

The f1-score, aka the harmonic mean of precision (how correct) and recall (how incorrect) is 1 for Setosa, 0.94 for Versicolor and 0.95 for Virginica

The average is 0.96
